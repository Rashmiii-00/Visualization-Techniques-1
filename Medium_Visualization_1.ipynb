```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

The data consist of linear values upto a specific range and its corresponding squares (number raised to power of 2) and cubes (number raised to power of 3). 

```python
dictionary = {
    "linear" : [i for i in range(1,101)],
    "square" : [i**2 for i in range(1,101)],
    "cube" : [i**3 for i in range(1,101)]
}
dictionary
```

```python
df = pd.DataFrame.from_dict(dictionary)
df
```

On Y-axis, we have Linear values upto 100, and on X-axis we have squares and cubes. The plot in Orange color represent Square and the plot in blue color represent Cube

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=df[:10], x='square', y='linear', label="Linear Vs Square")
sns.lineplot(data=df[:10], x='cube', y='linear', label="Linear Vs Cube")
plt.legend()
plt.title("Linear VS Square and Cube")
plt.show()
```

```python
random = np.random.randint(100, size=50)
```

```python
dictionary = {
    "random_integers" : [i for i in random],
    "square" : [i**2 for i in random],
    "cube" : [i**3 for i in random]
}
dictionary
```

```python
df1 = pd.DataFrame.from_dict(dictionary)
df1
```

```python
plt.figure(figsize=(20,8))
sns.scatterplot(data=df1, x='square', y='random_integers', label="Linear Vs Square")
sns.scatterplot(data=df1, x='cube', y='random_integers', label="Linear Vs Cube")
plt.title("Scatter plot for Linear Vs Square and Cube")
plt.legend()
plt.show()
```

```python
y = [i for i in range(50)]
sns.scatterplot(x=random, y=y)
plt.xlabel("Random Integers")
plt.ylabel("Linear Integers")
plt.title("Scatter plot for Random Integers")
```

```python
#line plot for random integers
sns.lineplot(x=random, y=y)
```
